```python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential
import dotenv

endpoint = "https://models.github.ai/inference"
model = "openai/gpt-5"
dotenv.load_dotenv()
token = os.getenv("TOKEN")
client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

response = client.complete(
    messages=[
        SystemMessage("You are a helpful assistant."),
        UserMessage("What is the capital of France?"),
    ],
    model=model
)

print(response.choices[0].message.content)

```

```python
from google import genai
import os
import dotenv
dotenv.load_dotenv()

# The client gets the API key from the environment variable `GEMINI_API_KEY`.
client = genai.Client(api_key=os.getenv("GEMINI_API_KEY"))

response = client.models.generate_content(
    model="gemini-2.5-flash", contents="when malaysia merdeka"
)
print(response.text)
```

```python
import urllib.parse

def generate_map_link(location_name):
    # Encode the string (replaces spaces with + and handles special characters)
    encoded_query = urllib.parse.quote_plus(location_name)
    return f"https://www.google.com/maps/search/?api=1&query={encoded_query}"

# Usage
location = "RUANG LOBI, TAMURA ELECTRONICS (M) SDN BHD, NO. 2, JLN HALBA 16/16, SEKSYEN 16, 40200 SHAH ALAM, SELANGOR"
print(generate_map_link(location))
# Output: https://www.google.com/maps/search/?api=1&query=Pusat+Darah+Negara
```
